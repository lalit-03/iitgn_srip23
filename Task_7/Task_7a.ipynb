```python
import jax.numpy as np
import matplotlib.pyplot as plt
```

```python
def prior(x):
  y = 6*x*((1-x*x)**2)
  return y

x = np.linspace(0,1,1000)
plt.plot(x, prior(x))
```

Calculation the Posterior Distribution:

bayes (1).jpg

```python
def posterior(x):
  y = (240240/67)*(x**9)*((1-x)**2)*((1-x**2)**2)
  return y

x = np.linspace(0,1,1000)
plt.plot(x, posterior(x), label='Posterior')
plt.plot(x, prior(x), label = 'Prior')
plt.legend()
```
